```python
import pandas as pd
#pip install openpyxl

#지역별 신재생 에너지 발전량 
df = pd.read_csv('../Data/국가_온실가스_배출현황.csv', index_col=0)

print(df.head())
```

```python
import matplotlib.pyplot as plt

df.loc['총배출량'].plot(marker='o')
plt.title('총 온실가스 배출량 (2013~2022)')
plt.ylabel('백만 톤 CO2eq.')
plt.xlabel('연도')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# 연도, 총배출량
years = np.array(df.columns, dtype=int).reshape(-1, 1)
values = np.array(df.loc['총배출량']).reshape(-1, 1)

model = LinearRegression()
model.fit(years, values)

# 2050년까지 예측
future_years = np.arange(2013, 2051).reshape(-1, 1)
future_preds = model.predict(future_years)
00
# 시각화
plt.plot(years, values, label='실제값', marker='o')
plt.plot(future_years, future_preds, label='선형 회귀 예측', linestyle='--')
plt.title('2050년까지 총 온실가스 배출량 예측')
plt.xlabel('연도')
plt.ylabel('백만 톤 CO2eq.')
plt.legend()
plt.grid(True)
plt.show()
```
